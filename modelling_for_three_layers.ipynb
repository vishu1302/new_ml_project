```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

from sklearn.neural_network import MLPRegressor
from sklearn.metrics import r2_score

from sklearn.metrics import mean_squared_error
from math import sqrt
```

```python
data=pd.read_excel('coalcombined_Fluidity_norm.xlsx')
data.describe()
```

```python
#Declaring Variables
x=data[['Ash', 'Volatile matter', 'Carbon', 'Hydrogen', 'Nitrogen', 'Sulphur',
       'Rp', 'Rw', 'Vitrinite', 'Liptinite', 'BI']]
y=data['log(Max Fluidity)']
```

```python
from sklearn.model_selection import train_test_split
 
data.columns
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=365)
```

```python
dict2={}
dict3={}
for i in range(1,17):
    for j in range(1,17):
        for k in range(1,17):
            mlp=MLPRegressor(hidden_layer_sizes=(i,j,k),max_iter=5000, activation='relu' , learning_rate='adaptive', learning_rate_init=0.1)
            mlp.fit(x_train, y_train)
            pred=mlp.predict(x_test)
            r2=r2_score(y_test, pred)
            mse = mean_squared_error(y_test, pred)
            rmse = sqrt(mse)
            dict2[i,j,k]=r2
            dict3[i,j,k]=rmse
```

```python
dict2
```

```python
df1=pd.DataFrame(columns=['position','r2','rmse'])
list1=dict2.keys()
#print(list1)

df1['position']=list1
df1['r2']=dict2.values()
df1['rmse']=dict3.values()
```
